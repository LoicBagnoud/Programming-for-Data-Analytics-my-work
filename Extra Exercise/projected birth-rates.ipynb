### 1. Importing packages we need.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Displaying the data

```python
df = pd.read_csv("projectedbirths-cso.csv")

print(df.head())
```

### 3. Getting a Plot (Lineplot)

```python
sns.lineplot(data=df, x="Year", y="VALUE", hue="Sex")
plt.title("Projected birth rates by year - Simple Lineplot")
plt.xlabel("Number of Births")
plt.ylabel("Years")
plt.show()
```
