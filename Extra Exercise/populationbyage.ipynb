### 1. Importing the packages we need

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Transfering the data over

```python
df = pd.read_csv("cso-populationbyage.csv")
df["Administrative Counties"] = (df["Administrative Counties"].str.replace(" County Council", "", regex=False).str.replace(" City Council", "City", regex=False))
df.columns = df.columns.str.strip()
print(df.head())
```

### 3. Making the plots

```python
df_2022 = df[df["CensusYear"] == 2022]
df_2022 = df_2022[df_2022["Administrative Counties"] != "Ireland"]
df_2022 = df_2022[df_2022["Single Year of Age"] == "All ages"]

plt.figure(figsize=(18,18))
sns.barplot(data=df_2022, x="Administrative Counties", y="VALUE")
plt.xticks(rotation=75, fontsize=10)
plt.yticks(fontsize=10)
plt.title("Population by County (2022)", fontsize=14)
plt.show()
```
